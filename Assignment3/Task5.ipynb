```python
import gpt_2_simple as gpt2
import nltk
```

```python
from datetime import datetime
```

```python
gpt2.download_gpt2()
```

```python
sess = gpt2.start_tf_sess()
```

```python
file = open("Text1.txt","r")
if(file.mode == "r"):
    text1 = file.read()
file.close()

file = open("Text2.txt","r")
if(file.mode == "r"):
    text2 = file.read()
file.close()

file = open("Text3.txt","r")
if(file.mode == "r"):
    text3 = file.read()
file.close()
```

```python
#Concatenating the 3 texts
data = text1+text2+text3

#Converting upper case to lower case
data = data.lower()
```

```python
sentence_start_token = "SENTENCE_START"
sentence_end_token = "SENTENCE_END"
```

```python
file = open("Processed.txt", "a")
file.write(data)
file.close()

file = open("Processed.txt","r")
if(file.mode == "r"):
    Processed = file.read()
file.close()
```

```python
file_name='Processed.txt'
```

```python
gpt2.finetune(sess,
             dataset=file_name,
             model_name='124M',
             steps=100,
             restore_from='fresh',
             run_name='run1',
             print_every=10,
             sample_every=20,
             save_every=50
             )
```

```python
gpt2.generate(sess, run_name='run1',
             length=100,
             temperature=.7,
             nsamples=10,
             batch_size=10,
             prefix=sentence_start_token,
             truncate=sentence_end_token,
             include_prefix=True)
```
